```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import roc_auc_score, f1_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart.csv')
```

### EDA

```python
data['target'].value_counts(normalize=True)
```

```python
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numeric_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
fig, axs = plt.subplots(nrows=len(categorical_features), ncols=1, figsize=(8, 40))

for index, feature in enumerate(categorical_features):
    sns.countplot(x=feature, hue='target', data=data, ax=axs[index]);
```

```python
fig, axs = plt.subplots(nrows=len(numeric_features), ncols=1, figsize=(8, 40))

for index, feature in enumerate(numeric_features):
    sns.boxplot(x='target', y=feature, data=data, ax=axs[index]);
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(data[numeric_features].corr());
```
