## Calculate HOG features and detect face on image

```python
face_cascade = cv2.CascadeClassifier('/kaggle/input/haarcascade/haarcascade_frontalface_default.xml')

def detect_image_voila_jones(image, scaleFactor, minNeighbors, face_cascade=face_cascade):
    if not isinstance(image, type(None)):
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        gray = np.array(gray, dtype='uint8')

        faces_boundaries = face_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, 
                                                         minNeighbors=minNeighbors)
        detected_faces = []

        for (x,y,w,h) in faces_boundaries:
            img = cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
            detected_faces.append(image[y:y+h, x:x+w])

        return detected_faces
    else:
        return None
```

```python
def detect_image_hog(image, visualize=True):
    if not isinstance(image, type(None)):
        if len(image):
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            hog_data = hog(gray, orientations = 9, pixels_per_cell = (8, 8),
                           cells_per_block = (2, 2), feature_vector=True)
            if visualize:
                hog_features, hog_image = hog_data
                imshow(hog_image, cmap=plt.cm.gray);
            else:
                hog_features = hog_data
        else:
            return None
        return hog_features
    else:
        return None
```

## Viola-Jones + Hog detector

### Detect faces by Viola-Jones Face detector

```python
train_images_generator = (image.img_to_array(image.load_img(path, target_size=(256, 256)))
                          for path in train_images.image_name.values)

test_images_generator = (image.img_to_array(image.load_img(path, target_size=(256, 256)))
                          for path in test_images.image_name.values)
```

```python
%%time
viola_jones_detected_train = [[[detect_image_voila_jones(img, scale_factor, min_neighbours)
                                for img in train_images_generator] 
                               for scale_factor in range(3, 6)] 
                              for min_neighbours in range(3, 6)]
```

```python
viola_jones_detected_train = [[[img[0] if img else None 
                               for img in viola_jones_detected_train[i][j]
                              ] for j in range(3)] for i in range(3)]
```

```python
%%time
viola_jones_detected_test = [[[detect_image_voila_jones(img, scale_factor, min_neighbours)
                                for img in test_images_generator] 
                               for scale_factor in range(3, 6)] 
                              for min_neighbours in range(3, 6)]
```

```python
viola_jones_detected_test = [[[img[0] if img else None 
                               for img in viola_jones_detected_test[i][j]
                              ] for j in range(3)] for i in range(3)]
```

### Generate Hog embeddings for detected faces

```python
%%time
embeddings_train = [[[detect_image_hog(img, False)
                      for img in viola_jones_detected_train[i][j]]
                     for i in range(3)] 
                    for j in range(3)]
```

```python
%%time
embeddings_test = [[[detect_image_hog(img, False)
                      for img in viola_jones_detected_test[i][j]]
                     for i in range(3)] 
                    for j in range(3)]
```

```python
embeddings_train = [[[emb if not isinstance(emb, type(None)) else [] 
                      for emb in embeddings_train[i][j]]
                     for i in range(3)] 
                    for j in range(3)]
```

```python
embeddings_test = [[[emb if not isinstance(emb, type(None)) else [] 
                      for emb in embeddings_test[i][j]]
                     for i in range(3)] 
                    for j in range(3)]
```

#### Create train and test dataframes

```python
for i, j in itertools.product(range(3), range(3)):
    tmp_df = train_images[['image_name', 'target']]
    tmp_df = tmp_df.join(pd.DataFrame(embeddings_train[i][j]))
    tmp_df.to_csv('viola_jones_{}_{}_hog_embeddings_train.csv'.format(i+3, j+3), index=False)
```

```python
for i, j in itertools.product(range(3), range(3)):
    tmp_df = test_images[['image_name', 'target']]
    tmp_df = tmp_df.join(pd.DataFrame(embeddings_test[i][j]))
    tmp_df.to_csv('viola_jones_{}_{}_hog_embeddings_test.csv'.format(i+3, j+3), index=False)
```
